# 1. Descrição do Problema

Uma instituição de ensino (escola) necessita de um sistema online para gerenciar suas atividades acadêmicas e administrativas, facilitando a interação entre alunos, professores e a administração. A escola oferece diversos cursos e disciplinas em diferentes níveis de ensino.

Este trabalho visa criar uma base de dados para armazenar as informações envolvidas nas atividades da escola. Dentre essas informações está a descrição dos Cursos oferecidos, que deve estar identificada por pelo menos um nome, um código único e o departamento ao qual pertence. Cada curso é classificado por um nível de ensino (ex: fundamental, médio, técnico, graduação) e sua descrição deve incluir a carga horária total, o número de vagas, e pode indicar ainda a ementa, os pré-requisitos (outros cursos ou disciplinas) e a sala de aula padrão onde é ministrado.

Cada Disciplina que compõe um curso deve indicar quantas aulas semanais ela possui e o material didático básico recomendado. Cada curso pode ter diversas Regras (ex: frequência mínima obrigatória, critérios de aprovação) e diversas Necessidades de Infraestrutura (ex: laboratório de informática, projetor, lousa digital).

A escola contempla três tipos principais de Usuários: o Professor, que é quem ministra as disciplinas; o Aluno, que se matricula nas disciplinas; e o Funcionário Administrativo, que gerencia as operações. Todos os usuários têm um nome, sobrenome, data de nascimento, endereço, sexo, número de telefone, e-mail e senha para acesso ao sistema. Assuma que não existem dois usuários com o mesmo nome, sobrenome e número de telefone.

Cada Matrícula é feita por um aluno em uma ou mais disciplinas de um curso para um determinado período letivo (ex: semestre ou ano). Cada matrícula deve incluir também uma data da matrícula (quando foi efetivada), o status da matrícula (ex: ativa, trancada, concluída, reprovada) e as notas obtidas pelo aluno naquela disciplina. Pode incluir também informações sobre bolsas de estudo ou descontos aplicados. Note-se que cada matrícula em uma disciplina se refere a um único aluno e a uma oferta de disciplina específica. Uma matrícula pode ser confirmada ou cancelada até uma data limite antes do início do período letivo, e se confirmada, pode envolver o pagamento de taxas pelo aluno.

Cada disciplina pode ter muitas matrículas, mas o número de alunos matriculados não pode exceder a capacidade da turma ou da sala de aula designada. Cada disciplina é de responsabilidade de um ou mais professores, e cada professor pode ter informações sobre sua área de especialização e titulação. A escola também mantém um registro de Departamentos Acadêmicos, onde cada departamento possui um nome, um código único e um professor designado como chefe do departamento. Um departamento só pode ter um chefe.

Além disso, cada curso, disciplina, aluno e professor estão associados a uma Unidade da Escola (campus ou localidade específica), que corresponde à cidade, estado e país, e possivelmente a um prédio ou bloco específico dentro da unidade.

A comunicação no sistema é feita por meio de Avisos e Mensagens. Um professor pode enviar  ou mais mensagens para seus alunos (individualmente ou para a turma) e vice-versa. A administração também pode enviar avisos gerais. Cada mensagem é identificada por quem a enviou, o destinatário (ou grupo de destinatários) e o timestamp de sua criação, e deve incluir o texto da mensagem.

Um aluno pode Avaliar zero ou mais disciplinas e/ou professores ao final de um período letivo. Cada avaliação pode incluir um comentário em texto, e classificações para didática do professor, material de apoio, relevância do conteúdo e infraestrutura da sala.

suposições aplicadas:

- entende-se como redundante um curso ter pré-requisito de disciplinas. modelamos de forma que um `curso` pode exigir outros cursos (e.g. médio exigir fundamental) e tem disciplinas. Suas disciplinas podem ter como pré-requisito outras disciplinas. Dessa forma, é possível encontrar os pré-requisitos `curso-disciplina` por meio de queries.

- para garantir flexibilidade, um curso pode ou não ter sala de aula padrão.

- optou-se por modelar inicialmente `regras` e `infraestrutura` como entidade-fraca, para garantir máxima flexibilidade. caso com o uso mostrem-se padrões, pode-se promover para entidade forte utilizando categorias reutilizáveis.

- generalização  `usuario` -> {`professor`, `aluno`, `funcionario`} é total (todo usuário é um dos três) e disjunta (um mesmo `user_id` não acumula dois papéis)

- ao mapear `matricula` como entidade-associativa, garantimos que um aluno pode no máximo participar de uma `oferta_disciplina`.

- status inicial da `matricula` modelado pela existência de `dt_confirmacao` e `dt_cancelamento`

- para garantir a consistência de `oferta_disciplina`, mapeamos como uma entidade-associativa do relacionamento triplo `sala-disciplina-periodo_letivo`.
  - uma disciplina em um periodo letivo pode abrir turmas em salas diferentes: `0..N sala`
  - uma sala em um perído letivo pode comportar multiplas disciplinas: `0..N disciplina`
  - uma disciplina em uma sala pode ser ofertada em mais de um período letivo: `0..N periodo_letivo`.
  para garantir que uma disciplina possa ser ofertada em uma sala em um período letivo em mais de um horário, criamos um identificador em oferta disciplina (para referir a turma).


- capacidade da `oferta_disciplina` é construida pela capacidade da `sala` e outras requisitos da turma.

- garantimos o respeito da capacidade da sala ou turma por trigger (ou outra regra de negócio prioritária) na hora da criação de uma nova `matricula`, consultando `oferta_disciplina(capacidade)`

- entende-se como duas relações diferentes `professor--disciplina` (professor responsabiliza por uma disciplina) e `professor--oferta-disciplina` (professor oferta uma disciplina em um ano).

- um `professor` pode ministrar nenhuma ou várias `oferta_disciplina` e se responsabilizar por nenhuma ou várias `disciplina`
- uma `oferta_disciplina` ou `disciplina` pode vir de vários professores (co-docência).

- cada `professor` é chefe de no máximo um `departamento`, mas todo `departamento` tem exatamente um `professor` como chefe.

- todo `departamento`, `professor`, `aluno`, `disciplina` e `funcionario_adm` é vinculado e exatamente uma `unidade_escolar`

- uma `unidade_escolar` é referente a um único prédio, em um único bloco

- entende-se implicitamente que cada `sala` pertence a uma `unidade`


- `professor` ou `aluno` podem enviar `mensagem` que podem se destinar a qualquer `usuario` (privado), `oferta_disciplina` (todos os alunos da disciplina).
- toda `mensagem` precisa ter exatamente um `professor` ou `aluno` como rementente (modelado com trigger).

- `funcionario_adm` podem enviar `aviso` que pode se destinar a  qualquer `usuario`(privado), `oferta_disciplina` (todos os alunos da disciplina), ou ainda ao `curso` inteiro.

- `avaliacao` precisa receber `aluno` como avaliador, e avalia exatamente um entre `professor` ou `disciplina`, mas somente um (regra de negócio)

  * Ch0(Usuario) = {usuario_id}
* **Chaves Candidatas (CK)**:
  * Chk(Usuario) = {(nome, sobrenome, telefone)}

## Exercício 5

A normalização é um processo essencial para garantir a consistência dos dados e minimizar redundâncias e anomalias de atualização. Agora, iremos analisar o esquema relacional obtido no exercício 3, verificando em que forma normal cada tabela se encontra. Após isso, procederemos à normalização até a forma normal de Boyce-Codd (BCNF)

---

1. Primeira Forma Normal (1FN):
  * Todas as tabelas resultantes do mapeamento(*Usuario, Aluno, Professor, FuncionarioAdm, Curso, NivelCurso, Sala, Disciplina, PeriodoLetivo, Departamento, UnidadeEscolar, Aviso, Mensagem, Avaliacao, RegraCurso, InfraestruturaCurso, OfertaDisciplina, Matricula, ClassificadoComo, PreRequisitoCurso, PreRequisitoDisciplina, Integra, ResponsavelPor, MinistraPor, AvaliaDisciplina, AvaliaProfessor*) já estão na 1FN.

  * **Justificativa**: Com o processo de mapeamento do MER para MRel nos garante a 1FN. Atributos compostos(como *endereco* em *Usuarios* e *classificacoes* em *Avaliacao*) foram decompostos em atributos atômicos e monovalorados (*cidade, estado, pais; didatica, material_apoio, relevancia, infraestrutura*). Atributos multivalorados, se identificados, seriam mapeados para novas tabelas separadas, garantindo que cada atributo na tabela original seja monovalorado.

---

2. Segunda Forma Normal (2FN):
  * Todas as tabelas com chave primária de atributo único(ex: *Usuario, Curso, NivelCurso, Sala, Disciplina, PeriodoLetivo, Departamento, UnidadeEscolar, Aviso, Mensagem, Avaliacao, Aluno, Professor, FuncionarioAdm, OfertaDisciplina, Matricula*) estão automaticamente na 2FN, pois não podem existir dependências funcionais parciais de atributos não chave em relação a PK.

  * Para as tabelas com chaves primárias compostas:
    * *RegraCurso*(PK: (<u>regra_id, curso_id</u>)): O atributo *definicao* depende totalmente da PK, logo está na 2FN.
    * *InfraestruturaCurso*(PK: (<u>infraestrutura_id, curso_id</u>)): Similar a RegraCurso, *definicao* depende totalmente da PK, assim também está na 2FN.
    * As tabelas de relacionamento M:N(*ClassificadoComo, PreRequisitoCurso, PreRequisitoDisciplina, Integra, ResponsavelPor, MinistraPor, AvaliaDisciplina, AvaliaProfessor*) não possuem atributos não chave, sendo as PKs formadas por FKs, então estão na 2FN.

---

3. Terceira Forma Normal (3FN):
  * Para uma relação estar na 3FN ela precisa estar na 2FN e não haver dependências transitivas de atributos não chave em relação a PK.
  * A maioria das tabelas, como *Aluno, Professor, FuncionarioAdm, NivelCurso, Sala, Disciplina, PeriodoLetivo, UnidadeEscolar, Aviso, Mensagem, Avaliacao, RegraCurso, InfraestruturaCurso, OfertaDisciplina, Matricula,* e as tabelas de relacionamento M:N, estão na 3FN, pois seus atributos não chave dependem diretamente da chave primária(ou não há atributos não chave).
  * A tabela *Usuario* possui uma potencial violação então necessita de noramlização para 3FN.
    * DFs para *Usuario*: *usuario_id -> nome, sobrenome, telefone, data_nascimento, sexo, email, senha, cidade, estado, pais.*
    * Consideramos a dependência: *(cidade, estado) -> pais.*
    * Como *usuario_id -> (cidade, estado)* e *(cidade, estado) -> pais,* então *pais* tem uma dependência transitiva em relação a *usuario_id*.
    * **Normalização para 3FN**(*Usuario*): Para resolver esta violação da 3FN, a tabela *Usuario* deve ser decomposta:
      * **Usuario** = {<u>usuario_id</u>, nome, sobrenome, telefone, data_nascimento, sexo, email, senha, cidade_id}
        * Chave Primária (PK): Ch0(Usuario) = {usuario_id}
        * Chaves Candidatas (CK): Chk(Usuario) = {(nome, sobrenome, telefone)}
        * Chave Estrangeira (FK): CE(Usuario) = {<(cidade_id), Localidade(localidade_id)>}
      * **Localidade** = {<u>localidade_id</u>, cidade, estado, pais}
        * Chave Primária (PK): Ch0(Localidade) = {localidade_id}
        * Chaves Candidatas (CK): Chk(Localidade) = {(cidade, estado)}
      * **Justificativa**: Com esta decomposição é eliminado a dependência transitiva, isolando a informação sobre cidades, estados e países em uma tabela separada, assim evitando redundância e anomalias de atualização.


---

4. Forma Normal de Boyce-Codd(BCNF):
  * Uma relação está em BCFN se, para toda dependência funcional não trivial A -> B, A é uma superchave. Isso se aplica a todas as chaves candidatas.
  * Revisão após 3FN para Usuario:
    * No esquema atualizado de Usuario e Localidade, todas as DFs têm uma superchave no lado esquedo (LHS), logo estas tabelas estão na BCNF.
  * Para o restante das tabelas onde PK é atributo único, ou PKs compostas onde todos os não chabe atributos dependem da PK completa e não há DFs onde LHS não é uma superchave, elas já estão na BCFN. Então não foram identificadas outras DFs que violassem a BCNF.
