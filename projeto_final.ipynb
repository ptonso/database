# 1. Descrição do Problema

Uma instituição de ensino (escola) necessita de um sistema online para gerenciar suas atividades acadêmicas e administrativas, facilitando a interação entre alunos, professores e a administração. A escola oferece diversos cursos e disciplinas em diferentes níveis de ensino.

Este trabalho visa criar uma base de dados para armazenar as informações envolvidas nas atividades da escola. Dentre essas informações está a descrição dos Cursos oferecidos, que deve estar identificada por pelo menos um nome, um código único e o departamento ao qual pertence. Cada curso é classificado por um nível de ensino (ex: fundamental, médio, técnico, graduação) e sua descrição deve incluir a carga horária total, o número de vagas, e pode indicar ainda a ementa, os pré-requisitos (outros cursos ou disciplinas) e a sala de aula padrão onde é ministrado.

Cada Disciplina que compõe um curso deve indicar quantas aulas semanais ela possui e o material didático básico recomendado. Cada curso pode ter diversas Regras (ex: frequência mínima obrigatória, critérios de aprovação) e diversas Necessidades de Infraestrutura (ex: laboratório de informática, projetor, lousa digital).

A escola contempla três tipos principais de Usuários: o Professor, que é quem ministra as disciplinas; o Aluno, que se matricula nas disciplinas; e o Funcionário Administrativo, que gerencia as operações. Todos os usuários têm um nome, sobrenome, data de nascimento, endereço, sexo, número de telefone, e-mail e senha para acesso ao sistema. Assuma que não existem dois usuários com o mesmo nome, sobrenome e número de telefone.

Cada Matrícula é feita por um aluno em uma ou mais disciplinas de um curso para um determinado período letivo (ex: semestre ou ano). Cada matrícula deve incluir também uma data da matrícula (quando foi efetivada), o status da matrícula (ex: ativa, trancada, concluída, reprovada) e as notas obtidas pelo aluno naquela disciplina. Pode incluir também informações sobre bolsas de estudo ou descontos aplicados. Note-se que cada matrícula em uma disciplina se refere a um único aluno e a uma oferta de disciplina específica. Uma matrícula pode ser confirmada ou cancelada até uma data limite antes do início do período letivo, e se confirmada, pode envolver o pagamento de taxas pelo aluno.

Cada disciplina pode ter muitas matrículas, mas o número de alunos matriculados não pode exceder a capacidade da turma ou da sala de aula designada. Cada disciplina é de responsabilidade de um ou mais professores, e cada professor pode ter informações sobre sua área de especialização e titulação. A escola também mantém um registro de Departamentos Acadêmicos, onde cada departamento possui um nome, um código único e um professor designado como chefe do departamento. Um departamento só pode ter um chefe.

Além disso, cada curso, disciplina, aluno e professor estão associados a uma Unidade da Escola (campus ou localidade específica), que corresponde à cidade, estado e país, e possivelmente a um prédio ou bloco específico dentro da unidade.

A comunicação no sistema é feita por meio de Avisos e Mensagens. Um professor pode enviar  ou mais mensagens para seus alunos (individualmente ou para a turma) e vice-versa. A administração também pode enviar avisos gerais. Cada mensagem é identificada por quem a enviou, o destinatário (ou grupo de destinatários) e o timestamp de sua criação, e deve incluir o texto da mensagem.

Um aluno pode Avaliar zero ou mais disciplinas e/ou professores ao final de um período letivo. Cada avaliação pode incluir um comentário em texto, e classificações para didática do professor, material de apoio, relevância do conteúdo e infraestrutura da sala.

suposições aplicadas:

- entende-se como redundante um curso ter pré-requisito de disciplinas. modelamos de forma que um `curso` pode exigir outros cursos (e.g. médio exigir fundamental) e tem disciplinas. Suas disciplinas podem ter como pré-requisito outras disciplinas. Dessa forma, é possível encontrar os pré-requisitos `curso-disciplina` por meio de queries.

- para garantir flexibilidade, um curso pode ou não ter sala de aula padrão.

- optou-se por modelar inicialmente `regras` e `infraestrutura` como entidade-fraca, para garantir máxima flexibilidade. caso com o uso mostrem-se padrões, pode-se promover para entidade forte utilizando categorias reutilizáveis.

- generalização  `usuario` -> {`professor`, `aluno`, `funcionario`} é total (todo usuário é um dos três) e disjunta (um mesmo `user_id` não acumula dois papéis)

- ao mapear `matricula` como entidade-associativa, garantimos que um aluno pode no máximo participar de uma `oferta_disciplina`.

- status inicial da `matricula` modelado pela existência de `dt_confirmacao` e `dt_cancelamento`

- para garantir a consistência de `oferta_disciplina`, mapeamos como uma entidade-associativa do relacionamento triplo `sala-disciplina-periodo_letivo`.
  - uma disciplina em um periodo letivo pode abrir turmas em salas diferentes: `0..N sala`
  - uma sala em um perído letivo pode comportar multiplas disciplinas: `0..N disciplina`
  - uma disciplina em uma sala pode ser ofertada em mais de um período letivo: `0..N periodo_letivo`.
  para garantir que uma disciplina possa ser ofertada em uma sala em um período letivo em mais de um horário, criamos um identificador em oferta disciplina (para referir a turma).


- capacidade da `oferta_disciplina` é construida pela capacidade da `sala` e outras requisitos da turma.

- garantimos o respeito da capacidade da sala ou turma por trigger (ou outra regra de negócio prioritária) na hora da criação de uma nova `matricula`, consultando `oferta_disciplina(capacidade)`

- entende-se como duas relações diferentes `professor--disciplina` (professor responsabiliza por uma disciplina) e `professor--oferta-disciplina` (professor oferta uma disciplina em um ano).

- um `professor` pode ministrar nenhuma ou várias `oferta_disciplina` e se responsabilizar por nenhuma ou várias `disciplina`
- uma `oferta_disciplina` ou `disciplina` pode vir de vários professores (co-docência).

- cada `professor` é chefe de no máximo um `departamento`, mas todo `departamento` tem exatamente um `professor` como chefe.

- todo `departamento`, `professor`, `aluno`, `disciplina` e `funcionario_adm` é vinculado e exatamente uma `unidade_escolar`

- uma `unidade_escolar` é referente a um único prédio, em um único bloco

- entende-se implicitamente que cada `sala` pertence a uma `unidade`


- `professor` ou `aluno` podem enviar `mensagem` que podem se destinar a qualquer `usuario` (privado), `oferta_disciplina` (todos os alunos da disciplina).
- toda `mensagem` precisa ter exatamente um `professor` ou `aluno` como rementente (modelado com trigger).

- `funcionario_adm` podem enviar `aviso` que pode se destinar a  qualquer `usuario`(privado), `oferta_disciplina` (todos os alunos da disciplina), ou ainda ao `curso` inteiro.

- `avaliacao` precisa receber `aluno` como avaliador, e avalia exatamente um entre `professor` ou `disciplina`, mas somente um (regra de negócio)

  * Ch0(Usuario) = {usuario_id}
* **Chaves Candidatas (CK)**:
  * Chk(Usuario) = {(nome, sobrenome, telefone)}

```python

-- # TABELAS DE DOMÍNIO E ENTIDADES PRINCIPAIS (SEM DEPENDÊNCIAS EXTERNAS)


-- Tabela para armazenar localidades (cidades, estados)
CREATE TABLE Localidade (
    cidade VARCHAR(150) NOT NULL,
    estado VARCHAR(150) NOT NULL,
    pais VARCHAR(100) NOT NULL,
    CONSTRAINT localidade_pk PRIMARY KEY (cidade, estado)
);

-- Tabela para unidades escolares (campi, prédios)
CREATE TABLE UnidadeEscolar (
    unidade_escolar_id SERIAL PRIMARY KEY,
    cidade VARCHAR(150) NOT NULL,
    estado VARCHAR(150) NOT NULL,
    pais VARCHAR(100) NOT NULL,
    predio VARCHAR(100),
    bloco VARCHAR(50),
    CONSTRAINT unidade_escolar_localidade_fk FOREIGN KEY (cidade, estado) REFERENCES Localidade(cidade, estado) ON UPDATE CASCADE ON DELETE RESTRICT
);

-- Tabela base para todos os usuários do sistema
CREATE TABLE Usuario (
    usuario_id SERIAL PRIMARY KEY,
    nome VARCHAR(255) NOT NULL,
    sobrenome VARCHAR(255) NOT NULL,
    telefone VARCHAR(25) NOT NULL,
    data_nascimento DATE NOT NULL,
    sexo VARCHAR(15),
    email VARCHAR(255) UNIQUE NOT NULL,
    senha VARCHAR(255) NOT NULL,
    cidade VARCHAR(150),
    estado VARCHAR(150),
    pais VARCHAR(100),
    CONSTRAINT usuario_localidade_fk FOREIGN KEY (cidade, estado) REFERENCES Localidade(cidade, estado) ON UPDATE CASCADE ON DELETE SET NULL,
    CONSTRAINT usuario_nome_sobrenome_tel_ck UNIQUE (nome, sobrenome, telefone)
);

-- Tabela para os níveis de curso (ex: Graduação, Pós-Graduação)
CREATE TABLE NivelCurso (
    nivel_curso_id SERIAL PRIMARY KEY,
    nivel VARCHAR(100) NOT NULL
);

-- Tabela para períodos letivos (ex: 2024/1, 2024/2)
CREATE TABLE PeriodoLetivo (
    periodo_letivo_id SERIAL PRIMARY KEY,
    ano INTEGER NOT NULL,
    tipo VARCHAR(50) NOT NULL, -- Ex: '1º Semestre', 'Anual'
    data_inicio DATE,
    data_fim DATE,
    dt_limite_cancel DATE
);

-- Tabela para salas de aula
CREATE TABLE Sala (
    sala_id SERIAL PRIMARY KEY,
    capacidade INTEGER NOT NULL,
    unidade_escolar_id INTEGER NOT NULL,
    CONSTRAINT sala_unidade_escolar_fk FOREIGN KEY (unidade_escolar_id) REFERENCES UnidadeEscolar(unidade_escolar_id) ON UPDATE CASCADE ON DELETE CASCADE
);

-- Tabela para disciplinas
CREATE TABLE Disciplina (
    disciplina_id SERIAL PRIMARY KEY,
    codigo_nome VARCHAR(255) NOT NULL,
    aulas_semanais INTEGER,
    material_didatico TEXT,
    unidade_escolar_id INTEGER NOT NULL,
    CONSTRAINT disciplina_unidade_escolar_fk FOREIGN KEY (unidade_escolar_id) REFERENCES UnidadeEscolar(unidade_escolar_id) ON UPDATE CASCADE ON DELETE RESTRICT
);



-- # TABELAS DE ESPECIALIZAÇÃO E DEPENDENTES


-- Especialização de Usuario para Aluno
CREATE TABLE Aluno (
    usuario_id INTEGER PRIMARY KEY,
    nusp VARCHAR(20) UNIQUE,
    ano_ingresso INTEGER,
    unidade_escolar_id INTEGER,
    CONSTRAINT aluno_usuario_fk FOREIGN KEY (usuario_id) REFERENCES Usuario(usuario_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT aluno_unidade_escolar_fk FOREIGN KEY (unidade_escolar_id) REFERENCES UnidadeEscolar(unidade_escolar_id) ON UPDATE CASCADE ON DELETE SET NULL
);

-- Especialização de Usuario para Professor
CREATE TABLE Professor (
    usuario_id INTEGER PRIMARY KEY,
    titulacao VARCHAR(100),
    especializacao VARCHAR(255),
    unidade_escolar_id INTEGER,
    CONSTRAINT professor_usuario_fk FOREIGN KEY (usuario_id) REFERENCES Usuario(usuario_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT professor_unidade_escolar_fk FOREIGN KEY (unidade_escolar_id) REFERENCES UnidadeEscolar(unidade_escolar_id) ON UPDATE CASCADE ON DELETE SET NULL
);

-- Especialização de Usuario para Funcionário Administrativo
CREATE TABLE FuncionarioAdm (
    usuario_id INTEGER PRIMARY KEY,
    unidade_escolar_id INTEGER,
    CONSTRAINT funcionarioadm_usuario_fk FOREIGN KEY (usuario_id) REFERENCES Usuario(usuario_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT funcionarioadm_unidade_escolar_fk FOREIGN KEY (unidade_escolar_id) REFERENCES UnidadeEscolar(unidade_escolar_id) ON UPDATE CASCADE ON DELETE SET NULL
);

-- Tabela para Departamentos
CREATE TABLE Departamento (
    departamento_id SERIAL PRIMARY KEY,
    nome VARCHAR(255) UNIQUE NOT NULL,
    chefe_professor_id INTEGER UNIQUE,
    unidade_escolar_id INTEGER NOT NULL,
    CONSTRAINT departamento_chefe_fk FOREIGN KEY (chefe_professor_id) REFERENCES Professor(usuario_id) ON UPDATE CASCADE ON DELETE SET NULL,
    CONSTRAINT departamento_unidade_escolar_fk FOREIGN KEY (unidade_escolar_id) REFERENCES UnidadeEscolar(unidade_escolar_id) ON UPDATE CASCADE ON DELETE RESTRICT
);

-- Tabela para Cursos
CREATE TABLE Curso (
    codigo_id SERIAL PRIMARY KEY,
    nome VARCHAR(255) NOT NULL,
    ch_total INTEGER, -- Carga Horária Total
    numero_vagas INTEGER,
    ementa TEXT,
    departamento_id INTEGER,
    unidade_escolar_id INTEGER,
    CONSTRAINT curso_departamento_fk FOREIGN KEY (departamento_id) REFERENCES Departamento(departamento_id) ON UPDATE CASCADE ON DELETE SET NULL,
    CONSTRAINT curso_unidade_escolar_fk FOREIGN KEY (unidade_escolar_id) REFERENCES UnidadeEscolar(unidade_escolar_id) ON UPDATE CASCADE ON DELETE RESTRICT
);

-- Tabela para ofertas de disciplinas em um período letivo
CREATE TABLE OfertaDisciplina (
    oferta_id SERIAL PRIMARY KEY,
    disciplina_id INTEGER NOT NULL,
    periodo_letivo_id INTEGER NOT NULL,
    sala_id INTEGER,
    horario VARCHAR(100),
    dias_sem VARCHAR(100),
    capacidade_turma INTEGER,
    CONSTRAINT oferta_disciplina_fk FOREIGN KEY (disciplina_id) REFERENCES Disciplina(disciplina_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT oferta_periodo_letivo_fk FOREIGN KEY (periodo_letivo_id) REFERENCES PeriodoLetivo(periodo_letivo_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT oferta_sala_fk FOREIGN KEY (sala_id) REFERENCES Sala(sala_id) ON UPDATE CASCADE ON DELETE SET NULL,
    CONSTRAINT oferta_disciplina_periodo_sala_ck UNIQUE (disciplina_id, periodo_letivo_id, sala_id)
);

-- Tabela de Avaliações (feitas por alunos)
CREATE TABLE Avaliacao (
    avaliacao_id SERIAL PRIMARY KEY,
    aluno_id INTEGER NOT NULL,
    comentario TEXT,
    didatica SMALLINT, -- Ex: nota de 1 a 5
    material_apoio SMALLINT,
    relevancia SMALLINT,
    infraestrutura SMALLINT,
    CONSTRAINT avaliacao_aluno_fk FOREIGN KEY (aluno_id) REFERENCES Aluno(usuario_id) ON UPDATE CASCADE ON DELETE CASCADE
);



-- # TABELAS DE RELACIONAMENTO (MUITOS-PARA-MUITOS E ENTIDADES FRACAS)


-- Entidade Fraca: Regras de um Curso
CREATE TABLE RegraCurso (
    regra_id SERIAL,
    curso_id INTEGER,
    definicao TEXT NOT NULL,
    CONSTRAINT regracurso_pk PRIMARY KEY (curso_id, regra_id),
    CONSTRAINT regracurso_curso_fk FOREIGN KEY (curso_id) REFERENCES Curso(codigo_id) ON UPDATE CASCADE ON DELETE CASCADE
);

-- Entidade Fraca: Infraestrutura de um Curso
CREATE TABLE InfraestruturaCurso (
    infraestrutura_id SERIAL,
    curso_id INTEGER,
    definicao TEXT NOT NULL,
    CONSTRAINT infraestruturacurso_pk PRIMARY KEY (curso_id, infraestrutura_id),
    CONSTRAINT infraestruturacurso_curso_fk FOREIGN KEY (curso_id) REFERENCES Curso(codigo_id) ON UPDATE CASCADE ON DELETE CASCADE
);

-- Tabela de Matrícula (associa Aluno a uma Oferta de Disciplina)
CREATE TABLE Matricula (
    matricula_id SERIAL PRIMARY KEY,
    aluno_id INTEGER NOT NULL,
    oferta_disciplina_id INTEGER NOT NULL,
    data_matricula DATE NOT NULL DEFAULT CURRENT_DATE,
    status VARCHAR(50),
    nota_final NUMERIC(4, 2),
    bolsa_estudo BOOLEAN DEFAULT FALSE,
    desconto NUMERIC(5, 2),
    dt_confirmacao TIMESTAMP WITH TIME ZONE,
    dt_cancelamento TIMESTAMP WITH TIME ZONE,
    taxas NUMERIC(10, 2),
    CONSTRAINT matricula_aluno_fk FOREIGN KEY (aluno_id) REFERENCES Aluno(usuario_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT matricula_oferta_disciplina_fk FOREIGN KEY (oferta_disciplina_id) REFERENCES OfertaDisciplina(oferta_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT matricula_aluno_oferta_ck UNIQUE (aluno_id, oferta_disciplina_id)
);

-- Relacionamento M:N entre Curso e NivelCurso
CREATE TABLE ClassificadoComo (
    curso_id INTEGER NOT NULL,
    nivel_curso_id INTEGER NOT NULL,
    CONSTRAINT classificadacomo_pk PRIMARY KEY (curso_id, nivel_curso_id),
    CONSTRAINT classificadacomo_curso_fk FOREIGN KEY (curso_id) REFERENCES Curso(codigo_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT classificadacomo_nivel_fk FOREIGN KEY (nivel_curso_id) REFERENCES NivelCurso(nivel_curso_id) ON UPDATE CASCADE ON DELETE CASCADE
);

-- Auto-relacionamento M:N de pré-requisitos entre Cursos
CREATE TABLE PreRequisitoCurso (
    curso_id INTEGER NOT NULL,
    pre_req_curso_id INTEGER NOT NULL,
    CONSTRAINT prerequisitocurso_pk PRIMARY KEY (curso_id, pre_req_curso_id),
    CONSTRAINT prerequisitocurso_curso_fk FOREIGN KEY (curso_id) REFERENCES Curso(codigo_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT prerequisitocurso_prereq_fk FOREIGN KEY (pre_req_curso_id) REFERENCES Curso(codigo_id) ON UPDATE CASCADE ON DELETE CASCADE
);

-- Auto-relacionamento M:N de pré-requisitos entre Disciplinas
CREATE TABLE PreRequisitoDisciplina (
    disciplina_id INTEGER NOT NULL,
    pre_req_disciplina_id INTEGER NOT NULL,
    CONSTRAINT prerequisitodisciplina_pk PRIMARY KEY (disciplina_id, pre_req_disciplina_id),
    CONSTRAINT prerequisitodisciplina_disciplina_fk FOREIGN KEY (disciplina_id) REFERENCES Disciplina(disciplina_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT prerequisitodisciplina_prereq_fk FOREIGN KEY (pre_req_disciplina_id) REFERENCES Disciplina(disciplina_id) ON UPDATE CASCADE ON DELETE CASCADE
);

-- Relacionamento M:N entre Disciplina e Curso (Grade Curricular)
CREATE TABLE Integra (
    disciplina_id INTEGER NOT NULL,
    curso_id INTEGER NOT NULL,
    CONSTRAINT integra_pk PRIMARY KEY (disciplina_id, curso_id),
    CONSTRAINT integra_disciplina_fk FOREIGN KEY (disciplina_id) REFERENCES Disciplina(disciplina_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT integra_curso_fk FOREIGN KEY (curso_id) REFERENCES Curso(codigo_id) ON UPDATE CASCADE ON DELETE CASCADE
);

-- Relacionamento M:N entre Disciplina e Professor (Professores aptos a lecionar)
CREATE TABLE ResponsavelPor (
    disciplina_id INTEGER NOT NULL,
    professor_id INTEGER NOT NULL,
    CONSTRAINT responsavelpor_pk PRIMARY KEY (disciplina_id, professor_id),
    CONSTRAINT responsavelpor_disciplina_fk FOREIGN KEY (disciplina_id) REFERENCES Disciplina(disciplina_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT responsavelpor_professor_fk FOREIGN KEY (professor_id) REFERENCES Professor(usuario_id) ON UPDATE CASCADE ON DELETE CASCADE
);

-- Relacionamento M:N entre Professor e OfertaDisciplina (Professor que ministra a turma)
CREATE TABLE MinistraPor (
    professor_id INTEGER NOT NULL,
    oferta_disciplina_id INTEGER NOT NULL,
    CONSTRAINT ministrapor_pk PRIMARY KEY (professor_id, oferta_disciplina_id),
    CONSTRAINT ministrapor_professor_fk FOREIGN KEY (professor_id) REFERENCES Professor(usuario_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT ministrapor_oferta_disciplina_fk FOREIGN KEY (oferta_disciplina_id) REFERENCES OfertaDisciplina(oferta_id) ON UPDATE CASCADE ON DELETE CASCADE
);

-- Relacionamento M:N entre Avaliacao e Disciplina
CREATE TABLE AvaliaDisciplina (
    avaliacao_id INTEGER NOT NULL,
    disciplina_id INTEGER NOT NULL,
    CONSTRAINT avaliadisciplina_pk PRIMARY KEY (avaliacao_id, disciplina_id),
    CONSTRAINT avaliadisciplina_avaliacao_fk FOREIGN KEY (avaliacao_id) REFERENCES Avaliacao(avaliacao_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT avaliadisciplina_disciplina_fk FOREIGN KEY (disciplina_id) REFERENCES Disciplina(disciplina_id) ON UPDATE CASCADE ON DELETE CASCADE
);

-- Relacionamento M:N entre Avaliacao e Professor
CREATE TABLE AvaliaProfessor (
    avaliacao_id INTEGER NOT NULL,
    professor_id INTEGER NOT NULL,
    CONSTRAINT avaliaprofessor_pk PRIMARY KEY (avaliacao_id, professor_id),
    CONSTRAINT avaliaprofessor_avaliacao_fk FOREIGN KEY (avaliacao_id) REFERENCES Avaliacao(avaliacao_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT avaliaprofessor_professor_fk FOREIGN KEY (professor_id) REFERENCES Professor(usuario_id) ON UPDATE CASCADE ON DELETE CASCADE
);



-- # TABELAS DE COMUNICAÇÃO


-- Tabela para Avisos (enviados por funcionários)
CREATE TABLE Aviso (
    aviso_id SERIAL PRIMARY KEY,
    "timestamp" TIMESTAMP WITH TIME ZONE NOT NULL DEFAULT CURRENT_TIMESTAMP,
    mensagem TEXT NOT NULL,
    remetente_funcionario_adm_id INTEGER NOT NULL,
    destinatario_usuario_id INTEGER,
    destinatario_oferta_disciplina_id INTEGER,
    destinatario_curso_id INTEGER,
    CONSTRAINT aviso_remetente_fk FOREIGN KEY (remetente_funcionario_adm_id) REFERENCES FuncionarioAdm(usuario_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT aviso_dest_usuario_fk FOREIGN KEY (destinatario_usuario_id) REFERENCES Usuario(usuario_id) ON UPDATE CASCADE ON DELETE SET NULL,
    CONSTRAINT aviso_dest_oferta_fk FOREIGN KEY (destinatario_oferta_disciplina_id) REFERENCES OfertaDisciplina(oferta_id) ON UPDATE CASCADE ON DELETE SET NULL,
    CONSTRAINT aviso_dest_curso_fk FOREIGN KEY (destinatario_curso_id) REFERENCES Curso(codigo_id) ON UPDATE CASCADE ON DELETE SET NULL
);

-- Tabela para Mensagens diretas entre usuários
CREATE TABLE Mensagem (
    mensagem_id SERIAL PRIMARY KEY,
    "timestamp" TIMESTAMP WITH TIME ZONE NOT NULL DEFAULT CURRENT_TIMESTAMP,
    texto TEXT NOT NULL,
    remetente_usuario_id INTEGER NOT NULL,
    destinatario_usuario_id INTEGER,
    destinatario_oferta_disciplina_id INTEGER, -- Mensagem para uma turma
    CONSTRAINT mensagem_remetente_fk FOREIGN KEY (remetente_usuario_id) REFERENCES Usuario(usuario_id) ON UPDATE CASCADE ON DELETE CASCADE,
    CONSTRAINT mensagem_dest_usuario_fk FOREIGN KEY (destinatario_usuario_id) REFERENCES Usuario(usuario_id) ON UPDATE CASCADE ON DELETE SET NULL,
    CONSTRAINT mensagem_dest_oferta_fk FOREIGN KEY (destinatario_oferta_disciplina_id) REFERENCES OfertaDisciplina(oferta_id) ON UPDATE CASCADE ON DELETE SET NULL
);
```

## Exercício 5

A normalização é um processo essencial para garantir a consistência dos dados e minimizar redundâncias e anomalias de atualização. Agora, iremos analisar o esquema relacional obtido no exercício 3, verificando em que forma normal cada tabela se encontra. Após isso, procederemos à normalização até a forma normal de Boyce-Codd (BCNF)

---

1. Primeira Forma Normal (1FN):
  * Todas as tabelas resultantes do mapeamento(*Usuario, Aluno, Professor, FuncionarioAdm, Curso, NivelCurso, Sala, Disciplina, PeriodoLetivo, Departamento, UnidadeEscolar, Aviso, Mensagem, Avaliacao, RegraCurso, InfraestruturaCurso, OfertaDisciplina, Matricula, ClassificadoComo, PreRequisitoCurso, PreRequisitoDisciplina, Integra, ResponsavelPor, MinistraPor, AvaliaDisciplina, AvaliaProfessor*) já estão na 1FN.

  * **Justificativa**: Com o processo de mapeamento do MER para MRel nos garante a 1FN. Atributos compostos(como *endereco* em *Usuarios* e *classificacoes* em *Avaliacao*) foram decompostos em atributos atômicos e monovalorados (*cidade, estado, pais; didatica, material_apoio, relevancia, infraestrutura*). Atributos multivalorados, se identificados, seriam mapeados para novas tabelas separadas, garantindo que cada atributo na tabela original seja monovalorado.

---

2. Segunda Forma Normal (2FN):
  * Todas as tabelas com chave primária de atributo único(ex: *Usuario, Curso, NivelCurso, Sala, Disciplina, PeriodoLetivo, Departamento, UnidadeEscolar, Aviso, Mensagem, Avaliacao, Aluno, Professor, FuncionarioAdm, OfertaDisciplina, Matricula*) estão automaticamente na 2FN, pois não podem existir dependências funcionais parciais de atributos não chave em relação a PK.

  * Para as tabelas com chaves primárias compostas:
    * *RegraCurso*(PK: (<u>regra_id, curso_id</u>)): O atributo *definicao* depende totalmente da PK, logo está na 2FN.
    * *InfraestruturaCurso*(PK: (<u>infraestrutura_id, curso_id</u>)): Similar a RegraCurso, *definicao* depende totalmente da PK, assim também está na 2FN.
    * As tabelas de relacionamento M:N(*ClassificadoComo, PreRequisitoCurso, PreRequisitoDisciplina, Integra, ResponsavelPor, MinistraPor, AvaliaDisciplina, AvaliaProfessor*) não possuem atributos não chave, sendo as PKs formadas por FKs, então estão na 2FN.

---

3. Terceira Forma Normal (3FN):
  * Para uma relação estar na 3FN ela precisa estar na 2FN e não haver dependências transitivas de atributos não chave em relação a PK.
  * A maioria das tabelas, como *Aluno, Professor, FuncionarioAdm, NivelCurso, Sala, Disciplina, PeriodoLetivo, UnidadeEscolar, Aviso, Mensagem, Avaliacao, RegraCurso, InfraestruturaCurso, OfertaDisciplina, Matricula,* e as tabelas de relacionamento M:N, estão na 3FN, pois seus atributos não chave dependem diretamente da chave primária(ou não há atributos não chave).
  * A tabela *Usuario* possui uma potencial violação então necessita de noramlização para 3FN.
    * DFs para *Usuario*: *usuario_id -> nome, sobrenome, telefone, data_nascimento, sexo, email, senha, cidade, estado, pais.*
    * Consideramos a dependência: *(cidade, estado) -> pais.*
    * Como *usuario_id -> (cidade, estado)* e *(cidade, estado) -> pais,* então *pais* tem uma dependência transitiva em relação a *usuario_id*.
    * **Normalização para 3FN**(*Usuario*): Para resolver esta violação da 3FN, a tabela *Usuario* deve ser decomposta:
      * **Usuario** = {<u>usuario_id</u>, nome, sobrenome, telefone, data_nascimento, sexo, email, senha, cidade_id}
        * Chave Primária (PK): Ch0(Usuario) = {usuario_id}
        * Chaves Candidatas (CK): Chk(Usuario) = {(nome, sobrenome, telefone)}
        * Chave Estrangeira (FK): CE(Usuario) = {<(cidade_id), Localidade(localidade_id)>}
      * **Localidade** = {<u>localidade_id</u>, cidade, estado, pais}
        * Chave Primária (PK): Ch0(Localidade) = {localidade_id}
        * Chaves Candidatas (CK): Chk(Localidade) = {(cidade, estado)}
      * **Justificativa**: Com esta decomposição é eliminado a dependência transitiva, isolando a informação sobre cidades, estados e países em uma tabela separada, assim evitando redundância e anomalias de atualização.


---

4. Forma Normal de Boyce-Codd(BCNF):
  * Uma relação está em BCFN se, para toda dependência funcional não trivial A -> B, A é uma superchave. Isso se aplica a todas as chaves candidatas.
  * Revisão após 3FN para Usuario:
    * No esquema atualizado de Usuario e Localidade, todas as DFs têm uma superchave no lado esquedo (LHS), logo estas tabelas estão na BCNF.
  * Para o restante das tabelas onde PK é atributo único, ou PKs compostas onde todos os não chabe atributos dependem da PK completa e não há DFs onde LHS não é uma superchave, elas já estão na BCFN. Então não foram identificadas outras DFs que violassem a BCNF.
